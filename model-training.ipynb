```python
import supervision as sv
import numpy as np

def yolo_to_xyxy(bbox, img_width, img_height):
    """ Convert YOLO format [x_center, y_center, w, h] → [x1, y1, x2, y2] """
    x_center, y_center, w, h = bbox
    x1 = int((x_center - w / 2) * img_width)
    y1 = int((y_center - h / 2) * img_height)
    x2 = int((x_center + w / 2) * img_width)
    y2 = int((y_center + h / 2) * img_height)
    return x1, y1, x2, y2
```

```python
import cv2
import matplotlib.pyplot as plt
import supervision as sv
import glob
import os

# Load trained YOLO model
model = YOLO("runs/detect/train5/weights/best.pt")  # Ensure correct path

# Define test image paths
test_images = glob.glob("/kaggle/input/satelite/images/test/*.jpg")

# Define label folder path
gt_labels = "/kaggle/input/satelite/labels/test/test/"

# Process a few test images
for img_path in test_images[:4]:  # Select first 4 images
    img = cv2.imread(img_path)
    img_height, img_width = img.shape[:2]  # Get image dimensions

    # Get corresponding label file
    label_file = os.path.join(gt_labels, os.path.basename(img_path).replace(".jpg", ".txt"))

    # Draw ground truth boxes (blue)
    if os.path.exists(label_file):
        with open(label_file, "r") as f:
            for line in f.readlines():
                cls, x, y, w, h = map(float, line.split())
                x1, y1, x2, y2 = yolo_to_xyxy([x, y, w, h], img_width, img_height)
                img = cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Blue for GT

    # Draw predicted boxes (red)
    results = model(img_path)[0]
    for box in results.boxes.xyxy:
        x1, y1, x2, y2 = map(int, box)
        img = cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)  # Red for Predictions

    # Show image
    plt.figure(figsize=(6, 6))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```

```python
import os
import glob

# Path to ground truth labels
gt_labels_path = "/kaggle/input/satelite/labels/test/"

# Path to test images
test_images = glob.glob("/kaggle/input/satelite/images/test/*.jpg")

# Function to convert YOLO format to XYXY format
def yolo_to_xyxy(bbox, img_width, img_height):
    """ Convert YOLO format [x_center, y_center, w, h] → [x1, y1, x2, y2] """
    x_center, y_center, w, h = bbox
    x1 = int((x_center - w / 2) * img_width)
    y1 = int((y_center - h / 2) * img_height)
    x2 = int((x_center + w / 2) * img_width)
    y2 = int((y_center + h / 2) * img_height)
    return x1, y1, x2, y2

# Extract ground truth bounding boxes
gt_boxes = []

for img_path in test_images:
    img_name = os.path.basename(img_path).replace(".jpg", ".txt")  # Get corresponding label file
    label_file = os.path.join(gt_labels_path, img_name)

    if os.path.exists(label_file):  # If label file exists
        img_width, img_height = 416, 416  # Assuming all images are 416x416
        with open(label_file, "r") as f:
            for line in f.readlines():
                cls, x, y, w, h = map(float, line.split())
                gt_boxes.append(yolo_to_xyxy([x, y, w, h], img_width, img_height))
```

```python
import numpy as np
from supervision import Detections
from supervision.metrics import MeanAveragePrecision

# Initialize Supervision's mAP metric
metric = MeanAveragePrecision()

formatted_predictions = []
formatted_ground_truth = []

# Convert predictions into the correct format
for img_path in test_images:
    results = model(img_path)[0]  # Get YOLO results for the image
    
    # Convert bounding boxes to NumPy array
    pred_bboxes = results.boxes.xyxy.cpu().numpy() if results.boxes.xyxy is not None else np.array([])
    pred_scores = results.boxes.conf.cpu().numpy() if results.boxes.conf is not None else np.array([])
    pred_classes = results.boxes.cls.cpu().numpy() if results.boxes.cls is not None else np.array([])

    # Create Detections object
    detections = Detections(
        xyxy=pred_bboxes,
        confidence=pred_scores,
        class_id=pred_classes.astype(int)  # Ensure integer class IDs
    )
    formatted_predictions.append(detections)

# Convert ground truth into the correct format
for img_path in test_images:
    img_name = os.path.basename(img_path).replace(".jpg", ".txt")
    label_file = os.path.join(gt_labels_path, img_name)

    if os.path.exists(label_file):
        gt_bboxes = []
        gt_classes = []
        with open(label_file, "r") as f:
            for line in f.readlines():
                cls, x, y, w, h = map(float, line.split())
                x1, y1, x2, y2 = yolo_to_xyxy([x, y, w, h], 416, 416)  # Convert YOLO to XYXY
                gt_bboxes.append([x1, y1, x2, y2])
                gt_classes.append(int(cls))  # Ensure integer class IDs

        detections = Detections(
            xyxy=np.array(gt_bboxes),
            class_id=np.array(gt_classes)
        )
        formatted_ground_truth.append(detections)
```
